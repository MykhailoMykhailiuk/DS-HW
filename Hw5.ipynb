```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

from google.colab import drive
drive.mount('/content/drive')

file_path = '/content/drive/My Drive/data'

data = []

for folder in os.listdir(file_path):
  folder_name = os.path.join(file_path, folder)
  if not os.path.isdir(folder_name):
    continue
  for file in os.listdir(folder_name):
    file_name = os.path.join(folder_name, file)
    if not file.endswith('.csv'):
      continue
    df = pd.read_csv(file_name)
    x_max = df['accelerometer_X'].max()
    x_min = df['accelerometer_X'].min()
    y_max = df['accelerometer_Y'].max()
    y_min = df['accelerometer_Y'].min()
    z_max = df['accelerometer_Z'].max()
    z_min = df['accelerometer_Z'].min()
    x_mean = df['accelerometer_X'].mean()
    y_mean = df['accelerometer_Y'].mean()
    z_mean = df['accelerometer_Z'].mean()
    x_median = df['accelerometer_X'].median()
    y_median = df['accelerometer_Y'].median()
    z_median = df['accelerometer_Z'].median()
    x_std = df['accelerometer_X'].std()
    y_std = df['accelerometer_Y'].std()
    z_std = df['accelerometer_Z'].std()
    x_var = df['accelerometer_X'].var()
    y_var = df['accelerometer_Y'].var()
    z_var = df['accelerometer_Z'].var()

    data.append([x_max, x_min, x_mean, x_median, x_std, x_var, y_max, y_min, y_mean, y_median, y_std, y_var, z_max, z_min, z_mean, z_median, z_std, z_var, folder_name.split('/')[-1]])
    columns = ['x_max', 'x_min', 'x_mean', 'x_median', 'x_std', 'x_var', 'y_max', 'y_min', 'y_mean', 'y_median', 'y_std', 'y_var', 'z_max', 'z_min', 'z_mean', 'z_median', 'z_std', 'z_var', 'state']
    df = pd.DataFrame(data, columns=columns)
  df
```

  SVM

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report

X_train, X_test, y_train, y_test = train_test_split(df.drop('state', axis=1), df['state'], test_size=0.2, random_state=42)

svm_model = SVC().fit(X_train, y_train)

y_pred = svm_model.predict(X_test)

print(classification_report(y_test, y_pred))
```

  Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier().fit(X_train, y_train)

y_pred = rf_model.predict(X_test)

print(classification_report(y_test, y_pred))
```
