```python
#Вивести перші рядки таблиці за допомогою методу head
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8"
df = pd.read_html(url)[12]
df.head(5)
```

```python
#Визначте кількість рядків та стовпців у датафреймі (атрибут shape)
df.shape
```

```python
#Замініть у таблиці значення "—" на значення NaN
df.replace("—", np.nan, inplace=True)
df
```

```python
#Визначте типи всіх стовпців за допомогою dataframe.dtypes
df.dtypes
```

```python
#Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"
df = df.apply(pd.to_numeric, errors="ignore")
df.dtypes
```

```python
#Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
df.isnull().sum() / len(df)
```

```python
#Видаліть з таблиці дані по всій країні, останній рядок таблиці
df.iloc[: -1]
```

```python
#Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)
num_cols = df.select_dtypes(include=[np.number]).columns
df[num_cols] = df[num_cols]/10
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())
df
```

```python
#Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні
avg_uk = df["2019"].mean()
new_df = df[df["2019"] > avg_uk]
new_df
```

```python
#У якому регіоні була найвища народжуваність у 2014 році?
region = df[df["2014"] == df["2014"].max()]
region["Регіон"]
```

```python
#Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році
plt.figure(figsize=(10, 6))
plt.bar(df['Регіон'], df['2019'])
plt.xticks(rotation=45, ha='right')
plt.xlabel("Регіон")
plt.ylabel("Коефіцієнт народжуваності за 2019 рік")
plt.tight_layout()
plt.show()
```

```python
#Plot 1
plt.figure(figsize=(14, 7))
for region in df['Регіон']:
    plt.plot(df.columns[1:], df[df['Регіон'] == region].values.flatten()[1:], label=region)
plt.title('Зміна показників за роками для всіх регіонів')
plt.xlabel('Рік')
plt.ylabel('Показник')
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.grid(True)
plt.show()
```

```python
#Plot 2
plt.figure(figsize=(20, 7))
sns.heatmap(df.set_index('Регіон').T, cmap='coolwarm', annot=True, fmt=".2f")
plt.title('Теплова карта значень по регіонах та роках')
plt.xlabel('Регіон')
plt.ylabel('Рік')
plt.show()
```

```python
#Plot 3
selected_regions = ['Київ', 'Львівська', 'Донецька', 'Одеська']
plt.figure(figsize=(10, 6))
for region in selected_regions:
    plt.plot(df.columns[1:], df[df['Регіон'] == region].values.flatten()[1:], label=region)
plt.title('Зміна показників за роками для вибраних регіонів')
plt.xlabel('Рік')
plt.ylabel('Показник')
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.grid(True)
plt.show()
```
