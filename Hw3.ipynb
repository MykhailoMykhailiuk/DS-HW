```python
import numpy as np
import pandas as pd
```

```python
#напишіть функцію гіпотези лінійної регресії у векторному вигляді;

def hypothesis(X, w):
  return np.dot(X, w)
```

```python
#створіть функцію для обчислення функції втрат у векторному вигляді;
def loss_func(X, y, w):
  m = len(X)
  h = hypothesis(X, w)
  return (1/(2*m)) * np.sum(np.square(h-y))
```

```python
#реалізуйте один крок градієнтного спуску;
def grad_step(X, y, w, alpha):
  m = len(y)
  h = hypothesis(X, w)
  w -= (alpha/m) * np.dot(X.T, (h-y))
  return w
```

```python
#Градієнт;
def gradient_descent(X, y, alpha, num_iters):
  # Додавання стовпчика одиниць до матриці ознак для врахування вільного члена (bias)
  ones = np.ones((X.shape[0], 1))
  X = np.hstack((ones, X))

  #ініціалізація вагів
  w = np.zeros(X.shape[1])

  loss_history = []
  for _ in range(num_iters):
    w = grad_step(X, y, w, alpha)
    loss = loss_func(X, y, w)
    loss_history.append(loss)
  return w, loss_history
```

```python
#Нормалізація даних ціни, площі, кількості ванних кімнат та кількості спалень
df_norm = df.copy(deep=True)
df_norm['price'] = (df_norm['price'] - df_norm['price'].mean()) / df_norm['price'].std()
df_norm['area'] = (df_norm['area'] - df_norm['area'].mean()) / df_norm['area'].std()
df_norm['bathrooms'] = (df_norm['bathrooms'] - df_norm['bathrooms'].mean()) / df_norm['bathrooms'].std()
df_norm['bedrooms'] = (df_norm['bedrooms'] - df_norm['bedrooms'].mean()) / df_norm['bedrooms'].std()
df_norm
```

```python
#знайдіть найкращі параметри w для датасету прогнозуючу ціну на будинок залежно від площі, кількості ванних кімнат та кількості спалень;
X = df_norm[['area', 'bedrooms', 'bathrooms']].values
y = df_norm['price'].values
alpha = 0.01
num_iters = 10000
w, loss_history = gradient_descent(X, y, alpha, num_iters)
print(f'Gradient descent w: {w}')
```

```python
#знайдіть ці ж параметри за допомогою аналітичного рішення;
ones = np.ones((X.shape[0], 1))
X = np.hstack((ones, X))
w_analytical = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
print(f'Analytical w: {w_analytical}')
```
